```python
# Step 5: Extracting Segments with K-Means Clustering

# Prepare the data for clustering
features = data[['EV_Sales_Quantity']]

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Determine the optimal number of clusters using the elbow method
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(K)
plt.grid()
plt.show()

# Fit K-Means with the optimal number of clusters
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['Cluster'] = kmeans.fit_predict(features_scaled)

# Analyze the characteristics of each segment
segment_analysis = data.groupby('Cluster').agg({
    'EV_Sales_Quantity': ['mean', 'std', 'count']
}).reset_index()

segment_analysis.columns = ['Segment', 'Average EV Sales Quantity', 'EV Sales Std Dev', 'Count']
print("\nSegment Analysis:")
print(segment_analysis)

# Visualize the average EV Sales Quantity by Segment
plt.figure(figsize=(10, 6))
sns.barplot(data=segment_analysis, x='Segment', y='Average EV Sales Quantity', palette='viridis')
plt.title('Average EV Sales Quantity by Segment')
plt.xlabel('Segment')
plt.ylabel('Average EV Sales Quantity')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()

# Visualize the count of vehicles in each segment
plt.figure(figsize=(10, 6))
sns.barplot(data=segment_analysis, x='Segment', y='Count', palette='magma')
plt.title('Count of Vehicles by Segment')
plt.xlabel('Segment')
plt.ylabel('Count of Vehicles')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()

# Create a detailed description for each segment
segment_descriptions = {
    0: {"Name": "Low Sales Vehicles", "Description": "Vehicles with low sales quantities."},
    1: {"Name": "Moderate Sales Vehicles", "Description": "Vehicles with moderate sales quantities."},
    2: {"Name": "High Sales Vehicles", "Description": "Vehicles with high sales quantities."},
    3: {"Name": "Very High Sales Vehicles", "Description": "Vehicles with very high sales quantities."}
}

# Display the detailed descriptions for each segment
print("\nSegment Descriptions:")
for segment_id, details in segment_descriptions.items():
    print(f"Segment ID: {segment_id} - Name: {details['Name']} - Description: {details['Description']}")
```

```python
# Step 6: Profiling Segments

# Analyze segment characteristics
segment_profile = data.groupby('Vehicle_Type').agg({
    'EV_Sales_Quantity': ['mean', 'std', 'count'],
    'Year': lambda x: x.mode()[0],  # Most common year
}).reset_index()

segment_profile.columns = ['Vehicle Type', 'Average Sales', 'Sales Std Dev', 'Count', 'Most Common Year']
print("\nSegment Profile:")
print(segment_profile)


# Analyze sales trends over time for each vehicle type
sales_trends = data.groupby(['Year', 'Vehicle_Type']).agg({'EV_Sales_Quantity': 'sum'}).reset_index()

# Visualize sales trends over time
plt.figure(figsize=(12, 8))
sns.lineplot(data=sales_trends, x='Year', y='EV_Sales_Quantity', hue='Vehicle_Type', marker='o')
plt.title('Sales Trends Over Time by Vehicle Type')
plt.xlabel('Year')
plt.ylabel('Total EV Sales Quantity')
plt.xticks(rotation=45)
plt.grid()
plt.legend(title='Vehicle Type')
plt.show()

# Insights and Recommendations
print("\nInsights and Recommendations:")
for vehicle_type in segment_profile['Vehicle Type']:
    latest_sales = sales_trends[sales_trends['Vehicle_Type'] == vehicle_type]['EV_Sales_Quantity'].iloc[-1]
    print(f"Vehicle Type: {vehicle_type} - Latest Sales: {latest_sales}")
    if latest_sales < segment_profile[segment_profile['Vehicle Type'] == vehicle_type]['Average Sales'].values[0]:
        print(" - Consider targeted marketing campaigns to boost sales.")
    else:
        print(" - Continue leveraging successful strategies to maintain momentum.")
```
