```python
import pandas as pd

# Load the dataset
data = pd.read_csv('mcdonalds.csv')

# Display the first few rows of the dataset
print("Dataset Overview:")
print(data.head())

# Display the structure of the dataset
print("\nDataset Information:")
print(data.info())
```

```python
# Check unique values in key columns
print("\nUnique Values in 'Gender':")
print(data['Gender'].unique())

print("\nUnique Values in 'VisitFrequency':")
print(data['VisitFrequency'].unique())

print("\nSummary Statistics for 'Age':")
print(data['Age'].describe())  # Summary statistics for Age
```

```python
# Analyze preferences based on gender
gender_preferences = data.groupby('Gender').agg({
    'VisitFrequency': 'value_counts',
    'Age': ['mean', 'std']
}).reset_index()

print("\nGender Preferences Analysis:")
print(gender_preferences)

# Analyze preferences based on age groups
age_bins = [0, 18, 30, 40, 50, 60, 100]
age_labels = ['<18', '18-30', '31-40', '41-50', '51-60', '60+']
data['AgeGroup'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels)

age_group_preferences = data.groupby('AgeGroup').agg({
    'VisitFrequency': 'value_counts',
    'Gender': 'value_counts'
}).reset_index()

print("\nAge Group Preferences Analysis:")
print(age_group_preferences)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select relevant features for clustering
# For this example, we will use Age and Visit Frequency
features = data_cleaned[['Age', 'VisitFrequency']]

# Convert 'VisitFrequency' to numerical values for clustering
visit_frequency_mapping = {
    'Every three months': 1,
    'Once a month': 2,
    'Once a week': 3,
    'More than once a week': 4,
    'Once a year': 0,
    'Never': -1
}
features['VisitFrequency'] = features['VisitFrequency'].map(visit_frequency_mapping)

# Display the prepared features
print("\nPrepared Features for Clustering:")
print(features.head())
```

```python
# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Display the standardized features
print("\nStandardized Features:")
print(features_scaled[:5])
```

```python
# Determine the optimal number of clusters using the Elbow method
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(K)
plt.grid()
plt.show()
```

```python
# Based on the Elbow method, let's assume we choose k=3
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data_cleaned['Segment'] = kmeans.fit_predict(features_scaled)

# Display the segments assigned to each customer
print("\nSegments Assigned to Customers:")
print(data_cleaned[['Age', 'VisitFrequency', 'Segment']].head())
```

```python
# Analyze the characteristics of each segment
segment_analysis = data_cleaned.groupby('Segment').agg({
    'Age': ['mean', 'std', 'count'],
    'VisitFrequency': 'value_counts'
}).reset_index()

segment_analysis.columns = ['Segment', 'Average Age', 'Age Std Dev', 'Count', 'Visit Frequency Count']
print("\nSegment Analysis:")
print(segment_analysis)
```

```python
# Analyze the characteristics of each segment
segment_profiles = data_cleaned.groupby('Segment').agg({
    'Age': ['mean', 'std', 'count'],
    'VisitFrequency': lambda x: x.value_counts().idxmax()  # Most common visit frequency
}).reset_index()

# Rename the columns for clarity
segment_profiles.columns = ['Segment', 'Average Age', 'Age Std Dev', 'Customer Count', 'Most Common Visit Frequency']

# Display the segment profiles
print("\nSegment Profiles:")
print(segment_profiles)
```
